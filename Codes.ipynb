```python
# Compute Approximate Entorpy.

rand_small = np.random.rand(100)
rand_big = np.random.rand(1000)

# Define the function to compute approximate entropy.

def ApEn(U, m, r):
    """Compute Aproximate entropy"""
    def _maxdist(x_i, x_j):
        return max([abs(ua - va) for ua, va in zip(x_i, x_j)])

    def _phi(m):
        x = [[U[j] for j in range(i, i + m - 1 + 1)] for i in range(N - m + 1)]
        C = [len([1 for x_j in x if _maxdist(x_i, x_j) <= r]) / (N - m + 1.0) for x_i in x]
        return (N - m + 1.0)**(-1) * sum(np.log(C))

    N = len(U)
    return abs(_phi(m+1) - _phi(m))


# Print the approximate entropy of the data.
print('Approximate Entropy of the data: %f' % ApEn(df['Adj Close (BTC)'], 2, 0.2*np.std(df['Adj Close (BTC)']) ))

# Print the approximate entropy of the random data.
print('Approximate Entropy of the random small data: %f' % ApEn(rand_small, 2, 0.2*np.std(rand_small)))
print('Approximate Entropy of the random big data: %f' % ApEn(rand_big, 2, 0.2*np.std(rand_big)))

# Print the approximate entropy of the deasoonalized data.

print('Approximate Entropy of the deasoonalized data: %f' % ApEn(decompose_data['Adj Close (BTC)'], 2, 0.2*np.std(decompose_data['Adj Close (BTC)']) ))

```

```python
# Compute the sample entropy of the data.

def SampEn(U, m, r):
    """Compute Sample entropy"""
    def _maxdist(x_i, x_j):
        return max([abs(ua - va) for ua, va in zip(x_i, x_j)])

    def _phi(m):
        x = [[U[j] for j in range(i, i + m - 1 + 1)] for i in range(N - m + 1)]
        C = [len([1 for j in range(len(x)) if i != j and _maxdist(x[i], x[j]) <= r]) for i in range(len(x))]
        return sum(C)

    N = len(U)

    
    return -np.log(_phi(m+1) / _phi(m))

# print the sample entropy of the data.
print('Sample Entropy of the data: %f' % SampEn(df['Adj Close (BTC)'], 2, 0.2*np.std(df['Adj Close (BTC)'], ddof=1)))

# print the sample entropy of the random data.
print('Sample Entropy of the random small data: %f' % SampEn(rand_small, 2, 0.2*np.std(rand_small, ddof=1)))
print('Sample Entropy of the random big data: %f' % SampEn(rand_big, 2, 0.2*np.std(rand_big, ddof=1)))

# print the sample entropy of the deasoonalized data.
print('Sample Entropy of the deasoonalized data: %f' % SampEn(decompose_data['Adj Close (BTC)'], 2, 0.2*np.std(decompose_data['Adj Close (BTC)'], ddof=1)))

```

```python
# Smoothing the data.

# Define the function to smooth the data.
def smooth(x, window_len=11, window='hanning'):
    """smooth the data using a window with requested size.
    
    This method is based on the convolution of a scaled window with the signal.
    The signal is prepared by introducing reflected copies of the signal 
    (with the window size) in both ends so that transient parts are minimized
    in the begining and end part of the output signal.
    
    input:
        x: the input signal 
        window_len: the dimension of the smoothing window; should be an odd integer
        window: the type of window from 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'
            flat window will produce a moving average smoothing.
    output:
        the smoothed signal
        
    example:
        t = linspace(-2,2,0.1)
        x = sin(t)+randn(len(t))*0.1
        y = smooth(x)
        
    see also: 
    
    numpy.hanning, numpy.hamming, numpy.bartlett, numpy.blackman, numpy.conv
    scipy.signal.lfilter
    """
    if x.ndim != 1:
        raise ValueError("smooth only accepts 1 dimension arrays.")

    if x.size < window_len:
        raise ValueError("Input vector needs to be bigger than window size.")
    
    if window_len<3:
        return x
    
    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise ValueError("Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'")
    
    s = np.r_[x[window_len-1:0:-1], x, x[-1:-window_len:-1]]
    if window == 'flat': #moving average
        w = np.ones(window_len,'d')
    else:
        w = eval('np.' + window + '(window_len)')

    y = np.convolve(w/w.sum(), s, mode='valid')
    return y[window_len//2:-window_len//2+1]

# start the smoothing of the data.
smooth_data = smooth(df['Adj Close (BTC)'])

# Index the smoothed data.
smooth_data_index = pd.DataFrame(smooth_data, index=df.index)

# Rename the column.
smooth_data_index.columns = ['Adj Close (BTC)']



# Plot the smoothed data.
plt.figure(figsize=(12,8))
plt.plot(smooth_data_index['Adj Close (BTC)'], label='Smoothed data')
plt.plot(df['Adj Close (BTC)'], label='Original data')
plt.legend()
plt.show()
```
